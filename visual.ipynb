# Trực quan

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('CarPrice_Assignment.csv')
dataset.head()
```

```python
CompanyName = dataset['CarName'].apply(lambda x : x.split(' ')[0])
dataset.insert(3, 'CompanyName', CompanyName)
dataset.drop(['CarName'], axis = 1, inplace = True)
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
sns.distplot(dataset['wheelbase'], color = 'g')
plt.show()
```

```python
sns.distplot(dataset['curbweight'], color = 'g')
plt.show()
```

```python
sns.distplot(dataset['stroke'], color = 'g')
plt.show()
```

```python
sns.distplot(dataset['compressionratio'], color = 'g')
plt.show()
```

```python
plt.figure(figsize = (24, 6))
plt.subplot(1, 2, 1)
sns.boxplot(dataset['price'], color = 'g')
plt.grid(True)
plt.title('Boxplot For Outliers Price', fontweight = 'bold')

plt.subplot(1, 2, 2)
sns.distplot(dataset['price'], color = 'g')
plt.title('Car Price Distribution', fontweight = 'bold')
plt.grid(True)
plt.show()
```

```python
col_categorical = list(dataset.select_dtypes(include = 'object'))
col_categorical
```

```python
col_numeric = list(dataset.select_dtypes(exclude = 'object'))
col_numeric
```

```python
df_autox = pd.DataFrame(dataset.groupby(['CompanyName'])['price'].mean())
df_autox.plot.bar()
plt.gcf().set_size_inches(20, 6)
plt.title('Car Company Name vs Average Price')
plt.show()
```

```python
plt.figure(figsize = (30, 8))
sns.countplot(dataset['CompanyName'], palette = 'YlGnBu')
plt.show()
```

```python
plt.figure(figsize = (25, 7))
plt.subplot(1, 3, 1)
sns.boxplot(x = 'fueltype', y = 'price', data = dataset, palette = 'YlGnBu')
plt.subplot(1, 3, 2)
sns.boxplot(x = 'doornumber', y = 'price', data = dataset, palette = 'YlGnBu')
plt.subplot(1, 3, 3)
sns.boxplot(x = 'carbody', y = 'price', data = dataset, palette = 'YlGnBu')
plt.show()
```

```python
dataset.corr()
fig, ax = plt.subplots(figsize = (30, 15)) 
sns.heatmap(dataset.corr(), annot = True, cmap = 'YlGnBu')
ax.set_xticklabels(ax.get_xticklabels(), rotation = 0, horizontalalignment = 'right');
```

```python
fig, ax= plt.subplots(1, 2, figsize = (25, 8))

sns.distplot(dataset['price'], norm_hist = False, kde = False, ax = ax[0], color = 'b')
ax[0].set_xlabel('Car Price')
ax[0].set_ylabel('Count of cars',size = 12)
ax[0].set_title('Count Of Cars By Price', size = 15)

sns.distplot(dataset['price'], kde = True, ax = ax[1], color = 'g')
ax[1].set_xlabel('Car Price')
ax[1].set_ylabel('Relative Frequency of cars', size = 12)
ax[1].set_title('Density or Relative Frequency Of Cars By Price', size = 15)
```
