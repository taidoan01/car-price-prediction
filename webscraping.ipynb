```python
import requests
import re
import json
from bs4 import BeautifulSoup
import pandas as pd
from google.colab import files
```

Địa chỉ web cần crawl

```python
url = "https://www.cars-data.com/en/car-brands-cars-logos.html"
```

Tạo các thuộc tính dạng danh sách

```python
url_list = []
name_list = []
model_list = []
brand_list = []
price_list = []
eLabel_list = []
bodyType_list = []
length_list = []
height_list = []
width_list = []
weight_list = []

weightTotal_list = []
emissionsCO2_list = []
modelDate_list = []
fuelType_list = []
numberOfAxles_list = []
numberOfDoors_list = []
numberOfForwardGears_list = []

seatingCapacity_list = []
vehicleTransmission_list = []
cargoVolume_list = []
roofLoad_list = []
accelerationTime_list = []
driveWheelConfiguration_list = []

fuelCapacity_list = []
fuelConsumption_list = []
speed_list = []
payload_list = []
trailerWeight_list = []
vEengineType_list = []
vEfuelType_list = []

vEengineDisplacement_list = []
vEenginePower_list = []
torque_list = []
```

Hàm lấy thuộc tính

```python
def get_attributes(url, tag, attr):
	result = []
	match = []
	r = requests.get(url)
	soup = BeautifulSoup(r.content, 'html5lib')
	for _ in soup.find_all(tag, attrs = {'class': attr}):
		try:
			result+= re.findall(r'href="(.*?)" title=', str(_))
		except:
			pass
	for el in result:
		if "cars-data.com" in el:
			match.append(el)
	return list(set(match))
```

```python
def get_car(url):

	global url_list, name_list, model_list, brand_list, price_list, eLabel_list, bodyType_list, length_list, height_list, width_list, weight_list, weightTotal_list, emissionsCO2_list, modelDate_list, fuelType_list, numberOfAxles_list, numberOfDoors_list, numberOfForwardGears_list, seatingCapacity_list, vehicleTransmission_list, cargoVolume_list, roofLoad_list, accelerationTime_list, driveWheelConfiguration_list, fuelCapacity_list, fuelConsumption_list, speed_list, payload_list, trailerWeight_list, vEengineType_list, vEfuelType_list, vEengineDisplacement_list, vEenginePower_list, torque_list

	r = requests.get(url)
	soup = BeautifulSoup(r.content, 'html5lib')
	attr = soup.find_all('td', attrs = {'class': 'col-6'})

	attrs = {}
	for i in range(0, len(attr), 2):
		attrs[attr[i].text[:-1]] = attr[i+1].text
	
	soup = BeautifulSoup(r.content, 'html.parser')
	js = json.loads(soup.select_one("script[type='application/ld+json']").text)

	url = url
	name = js.get('name', 'n/a')
	model = js.get('brand', 'n/a')
	brand = js.get('manufacturer', 'n/a')
	price = attrs.get('Price', 'n/a')
	eLabel = attrs.get('Energy Label', 'n/a')
	bodyType = attrs.get('Body Type', 'n/a')
	length = attrs.get('Length', 'n/a')
	height = attrs.get('Height', 'n/a')
	width = attrs.get('Width', 'n/a')
	weight = attrs.get('Weight', 'n/a')

	weightTotal = js.get('weightTotal').get('value', 'n/a')
	emissionsCO2 = attrs.get('Co2 Emissions', 'n/a')
	modelDate = js.get('productionDate', 'n/a')
	fuelType = attrs.get('Fuel Type', 'n/a')
	numberOfAxles = js.get('numberOfAxles', 'n/a')
	numberOfDoors = js.get('numberOfDoors', 'n/a')
	numberOfForwardGears = js.get('numberOfForwardGears', 'n/a')

	seatingCapacity = js.get('vehicleSeatingCapacity', 'n/a')
	vehicleTransmission = js.get('vehicleTransmission', 'n/a')
	cargoVolume = attrs.get('Cargo Capacity', 'n/a')
	roofLoad = js.get('roofLoad').get('value', 'n/a')
	accelerationTime = js.get('accelerationTime').get('value', 'n/a')
	driveWheelConfiguration  = attrs.get('Drive Wheel ', 'n/a')

	fuelCapacity = attrs.get('Engine Capacity', 'n/a')
	fuelConsumption = attrs.get('Urban Consumption', 'n/a')
	speed = attrs.get('Top Speed', 'n/a')
	payload = attrs.get('Max. Payload', 'n/a')
	trailerWeight = attrs.get('Curb Weight', 'n/a')
	vEengineType = attrs.get('Engine/motor Type', 'n/a')
	vEfuelType = attrs.get('Fuel Type', 'n/a')

	vEengineDisplacement = attrs.get('Engine Capacity', 'n/a')
	vEenginePower = attrs.get('Engine/motor Type', 'n/a')
	torque = js.get('vehicleEngine')[0].get('torque').get('value', 'n/a')
	
	values = url + '\t' + name + '\t' + model + '\t' + brand + '\t' + price + '\t' + eLabel + '\t' + bodyType + '\t' + length + '\t' + height + '\t' + width + '\t' + weight + '\t' + weightTotal + '\t' + emissionsCO2 + '\t' + modelDate + '\t' + fuelType + '\t' + numberOfAxles + '\t' + numberOfDoors + '\t' + numberOfForwardGears + '\t' + seatingCapacity + '\t' + vehicleTransmission + '\t' + cargoVolume + '\t' + roofLoad + '\t' + accelerationTime + '\t' + driveWheelConfiguration + '\t' + fuelCapacity + '\t' + fuelConsumption + '\t' + speed + '\t' + payload + '\t' + trailerWeight + '\t' + vEengineType + '\t' + vEfuelType + '\t' + vEengineDisplacement + '\t' + vEenginePower + '\t' + torque
	

	url_list.append(url)
	name_list.append(name)
	model_list.append(model)
	brand_list.append(brand)
	price_list.append(price)
	eLabel_list.append(eLabel)
	bodyType_list.append(bodyType)
	length_list.append(length)
	height_list.append(height)
	width_list.append(width)
	weight_list.append(weight)

	weightTotal_list.append(weightTotal)
	emissionsCO2_list.append(emissionsCO2)
	modelDate_list.append(modelDate)
	fuelType_list.append(fuelType)
	numberOfAxles_list.append(numberOfAxles)
	numberOfDoors_list.append(numberOfDoors)
	numberOfForwardGears_list.append(numberOfForwardGears)

	seatingCapacity_list.append(seatingCapacity)
	vehicleTransmission_list.append(vehicleTransmission)
	cargoVolume_list.append(cargoVolume)
	roofLoad_list.append(roofLoad)
	accelerationTime_list.append(accelerationTime)
	driveWheelConfiguration_list.append(driveWheelConfiguration)

	fuelCapacity_list.append(fuelCapacity)
	fuelConsumption_list.append(fuelConsumption)
	speed_list.append(speed)
	payload_list.append(payload)
	trailerWeight_list.append(trailerWeight)
	vEengineType_list.append(vEengineType)
	vEfuelType_list.append(vEfuelType)

	vEengineDisplacement_list.append(vEengineDisplacement)
	vEenginePower_list.append(vEenginePower)
	torque_list.append(torque)
	return values
```

```python
manufacturers = get_attributes(url, 'div', 'col-2 center')
print('All manufacturers:', len(manufacturers))
print(manufacturers)
print('-----')
```

```python
print(manufacturers[:15])
```

```python
manufacturers = manufacturers[:15]
print('crawl 15 manufacturers')
model = []
for m in manufacturers:
	model+= get_attributes(m, 'div', 'col-4')

print('All models:', len(model))
print(model)
print('-----')

model_with_options = []

for m in model:
	model_with_options+= get_attributes(m, 'div', 'col-4')

print('All model_with_options:', len(model_with_options))
print(model_with_options)
print('-----')

model_types_all_years = []

for m in model_with_options:
	model_types_all_years+= get_attributes(m, 'a', 'typesallyears')

print('All model_types_all_years:', len(model_types_all_years))
print(model_types_all_years)
print('-----')

model_types = []
for m in model_types_all_years:
	model_types+= get_attributes(m, 'div', 'col-6')

print('All cars:', len(model_types))
print(model_types)
print('-----')

for car in model_types:
	print(get_car(car))
```
