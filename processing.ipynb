```python
import pandas as pd
```

Load Data thô 

```python
cars_data = pd.read_csv('cars_dataset_15_manufactures.csv',sep = ';',encoding='utf-8')
cars_data
```

```python
cars_data.columns
```

Loại bỏ những cột không cần thiết

```python
col_drop = {'url','model', 'eLabel',
            'bodyType','weight','emissionsCO2', 
            'driveWheelConfiguration','vEfuelType',
            'vEenginePower','fuelConsumption'}
cars_data = cars_data.drop(col_drop,axis = 1)
cars_data
```

Loại bỏ đơn vị của các thuộc tính dạng số 

```python
cars_data['length'] = cars_data['length'].map(lambda x: x.rstrip('mm'))
cars_data['height'] = cars_data['height'].map(lambda x: x.rstrip('mm'))
cars_data['width'] = cars_data['width'].map(lambda x: x.rstrip('mm'))
cars_data['fuelCapacity'] = cars_data['fuelCapacity'].str.replace(r'\D','')
cars_data['vEengineDisplacement'] = cars_data['vEengineDisplacement'].str.replace(r'\D','')
cars_data['cargoVolume'] = cars_data['cargoVolume'].map(lambda x: x.strip('l'))
```

Hàm để xử lý cột cargoVolume

```python
def cvtFloat_cargoVolume(x):
    temp = x.split()[0]
    temp = temp.replace('-', ' ')
    temp = temp.split()
    if len(temp) > 0:
        temp = temp[-1]
    else:
        temp = x
    val = None
    try:
        val = float(temp)
    except ValueError:
        return val
    return val
```

```python
cars_data['cargoVolume'] = cars_data['cargoVolume'].apply(cvtFloat_cargoVolume)
```

```python
cars_data['speed'] = cars_data['speed'].map(lambda x: x.rstrip('km/h'))
cars_data['payload'] = cars_data['payload'].map(lambda x: x.rstrip('kg'))
cars_data['trailerWeight'] = cars_data['trailerWeight'].map(lambda x: x.rstrip('kg'))
```

Hàm ép kiểu float

```python
def cvtFloat(x):
    if type(x) == str:
        temp = x.replace(',', '.').split()[0]
    else:
        temp = x
    val = None
    try:
        val = float(temp)
    except ValueError:
        return val
    return val
```

```python
cars_data['accelerationTime'] = cars_data['accelerationTime'].apply(cvtFloat)
```

```python
cars_data
```

Loại bỏ các dòng bị lỗi, các dòng không lấy được dữ liệu

```python
cars_data = cars_data[cars_data != '-']
cars_data
```

```python
cars_data = cars_data[cars_data != '#NAME?']
cars_data
```

```python
cars_data = cars_data[cars_data != 'n.b.']
cars_data
```

```python
cars_data = cars_data[cars_data != 't']
cars_data
```

```python
cars_data = cars_data.dropna()
cars_data
```

Lưu lại thành csv

```python
cars_data.to_csv('cars_data_processed.csv',sep = '\t')
```
